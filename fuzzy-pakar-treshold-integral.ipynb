### FUZZIFIKASI

Fungsi Keanggotaan Umum

```python
# fungsi keanggotaan segitiga
def keanggotaan_segitiga1(x, a, b, c):
    if x == b:
        return 1
    elif x <= a or x >= c:
        return 0
    elif a < x < b:
        return (x - a) / (b - a)
    elif b < x < c:
        return (c - x) / (c - b)

def keanggotaan_segitiga2(x, arr):
    a, b, c = arr
    if x == b:
        return 1
    elif x <= a or x >= c:
        return 0
    elif a < x < b:
        return (x - a) / (b - a)
    elif b < x < c:
        return (c - x) / (c - b)
```

Fungsi Keanggotaan tiap Variabel

```python
# domain waktu belajar
wb_kecil_a = 0
wb_kecil_b = 0
wb_kecil_c = 454

wb_sedang_a = 356
wb_sedang_b = 454
wb_sedang_c = 527

wb_besar_a = 454
wb_besar_b = 700
wb_besar_c = 700

# domain lama jawab soal
ljs_kecil_a = 0
ljs_kecil_b = 0
ljs_kecil_c = 621

ljs_sedang_a = 367
ljs_sedang_b = 621
ljs_sedang_c = 1056

ljs_besar_a = 621
ljs_besar_b = 7500
ljs_besar_c = 7500

# domain nilai
n_kecil_a = 0
n_kecil_b = 0
n_kecil_c = 83

n_sedang_a = 66.7
n_sedang_b = 83
n_sedang_c = 100

n_besar_a = 83
n_besar_b = 100
n_besar_c = 100

# representasi posisi partikel
posisi_partikel = [
    wb_kecil_c, wb_sedang_a, wb_sedang_b, wb_sedang_c, wb_besar_a,
    ljs_kecil_c, ljs_sedang_a, ljs_sedang_b, ljs_sedang_c, ljs_besar_a, 
    n_kecil_c, n_sedang_a, n_sedang_b, n_sedang_c, ljs_besar_a,
]

# fuzzifikasi variabel waktu_belajar
def fuzzifikasi_waktu_belajar(x, param):
    kecil = keanggotaan_segitiga1(x, wb_kecil_a, wb_kecil_b, param[0])
    sedang = keanggotaan_segitiga2(x, param[1:4])
    besar = keanggotaan_segitiga1(x, param[4], wb_besar_b, wb_besar_c)
    return {'kecil': kecil, 'sedang': sedang, 'besar': besar}

# fuzzifikasi variabel lama_jawab_soal
def fuzzifikasi_lama_jawab_soal(x, param):
    kecil = keanggotaan_segitiga1(x, ljs_kecil_a, ljs_kecil_b, param[0])
    sedang = keanggotaan_segitiga2(x, param[1:4])
    besar = keanggotaan_segitiga1(x, param[4], ljs_besar_b, ljs_besar_c)
    return {'kecil': kecil, 'sedang': sedang, 'besar': besar}

# fuzzifikasi variabel nilai
def fuzzifikasi_nilai(x, param):
    kecil = keanggotaan_segitiga1(x, n_kecil_a, n_kecil_b, param[0])
    sedang = keanggotaan_segitiga2(x, param[1:4])
    besar = keanggotaan_segitiga1(x, param[4], n_besar_b, n_besar_c)
    return {'kecil': kecil, 'sedang': sedang, 'besar': besar}

# fuzzifikasi setiap variabel
def func_fuzzifikasi(waktu_belajar, lama_jawab_soal, nilai, param):
    param_wb = param[:5]
    param_ljs = param[5:10]
    param_n = param[10:]
    x = fuzzifikasi_waktu_belajar(waktu_belajar, param_wb)
    y = fuzzifikasi_lama_jawab_soal(lama_jawab_soal, param_ljs)
    z = fuzzifikasi_nilai(nilai, param_n)
    return x, y, z

x, y, z = func_fuzzifikasi(590,452,75, posisi_partikel)
print(f"waktu belajar   : {x}")
print(f"lama jawab soal : {y}")
print(f"nilai           : {z}")
```

### INFERENSI MAMDANI

##### Implikasi setiap rule

```python
def inferensi_mamdani(x, y, z, rules):
    hasil_inferensi = []
    rule_text = []
    for i in range(len(rules)):
        waktu_belajar = x[rules.loc[i, 'rerata_durasi_akses_materi']]
        lama_jawab_soal = y[rules.loc[i, 'durasi_berlatih']]
        nilai = z[rules.loc[i, 'nilai']]
        # menghitung alpapredikat setiap rule dengan AND (min)
        alpapredikat = min(waktu_belajar, lama_jawab_soal, nilai)
        # menambahkan ke hasil_inferensi
        hasil_inferensi.append({'rule': (i + 1), 'kepercayaan_pakar': rules.loc[i, 'kepercayaan_pakar'], 'alpapredikat': alpapredikat})
        text = (
            f"RULE {i + 1}: IF waktu_belajar {rules.loc[i, 'rerata_durasi_akses_materi']} ({waktu_belajar:.4f}) "
            f"AND lama_jawab_soal {rules.loc[i, 'durasi_berlatih']} ({lama_jawab_soal:.4f}) "
            f"AND nilai {rules.loc[i, 'durasi_berlatih']} ({nilai:.4f}) "
            f"THEN kelulusan {rules.loc[i, 'kelulusan']} : ALPAPREDIKAT = {alpapredikat:.4f}"
        )
        rule_text.append(text)
    return hasil_inferensi, rule_text

hasil_inferensi, rule_text = inferensi_mamdani(x, y, z, rules)
for i in range(len(rule_text)):
    print(rule_text[i])
for i in range(len(hasil_inferensi)):
    print(hasil_inferensi[i])
```

##### Agregasi

```python
def agregasi(hasil_inferensi):
    agregasi_kelulusan = {'kecil': 0, 'sedang': 0, 'besar': 0}
    for hasil in hasil_inferensi:
        alpapredikat = hasil['alpapredikat']
        kepercayaan_pakar = hasil['kepercayaan_pakar']
        if kepercayaan_pakar == 'kecil':
            if alpapredikat > agregasi_kelulusan['kecil']:
                agregasi_kelulusan['kecil'] = alpapredikat
        elif kepercayaan_pakar == 'sedang':
            if alpapredikat > agregasi_kelulusan['sedang']:
                agregasi_kelulusan['sedang'] = alpapredikat
        else:
            if alpapredikat > agregasi_kelulusan['besar']:
                agregasi_kelulusan['besar'] = alpapredikat
    return agregasi_kelulusan

hasil_agregasi = agregasi(hasil_inferensi)
hasil_agregasi
```

### DEFUZZIFIKASI COA

```python
def defuzzifikasi(agregasi_kelulusan):
    range_kecil = [0, 25]
    range_sedang = [25, 70]
    range_besar = [75, 100]
    
    momen_kecil = agregasi_kelulusan['kecil'] * (((range_kecil[1] ** 2) / 2) - ((range_kecil[0] ** 2) / 2))
    momen_sedang = agregasi_kelulusan['sedang'] * (((range_sedang[1] ** 2) / 2) - ((range_sedang[0] ** 2) / 2))
    momen_besar = agregasi_kelulusan['besar'] * (((range_besar[1] ** 2) / 2) - ((range_besar[0] ** 2) / 2))
    
    luas_kecil = agregasi_kelulusan['kecil'] * (range_kecil[1] - range_kecil[0])
    luas_sedang = agregasi_kelulusan['sedang'] * (range_sedang[1] - range_sedang[0])
    luas_besar = agregasi_kelulusan['besar'] * (range_besar[1] - range_besar[0])

    total_momen = momen_kecil + momen_sedang + momen_besar
    total_luas = luas_kecil + luas_sedang + luas_besar
    if total_luas == 0:
        return 0
    else:
        return total_momen / total_luas

hasil_defuzzifikasi = defuzzifikasi(hasil_agregasi)
print("Hasil Defuzzifikasi: ", hasil_defuzzifikasi)
```

### TRESHOLD KELULUSAN

```python
def kelulusan(hasil_defuzzifikasi):
    if hasil_defuzzifikasi >= 65:
        return '1'
    else:
        return '0'
    
hasil_kelulusan = kelulusan(hasil_defuzzifikasi)
hasil_kelulusan
```

### CEK AKURASI

```python
prediksi_benar = 0

for i in range(len(dataset)):
    waktu_belajar = dataset.loc[i, 'rerata_durasi_akses_materi']
    lama_jawab_soal = dataset.loc[i, 'durasi_berlatih']
    nilai = dataset.loc[i, 'nilai']
    aktual_kelulusan = dataset.loc[i, 'kelulusan']
    # fuzzifikasi
    x, y, z = func_fuzzifikasi(waktu_belajar, lama_jawab_soal, nilai, posisi_partikel)
    # inferensi mamdani
    hasil_inferensi, rule_text = inferensi_mamdani(x, y, z, rules)
    # agregasi
    hasil_agregasi = agregasi(hasil_inferensi)
    # defuzzifikasi coa
    hasil_defuzzifikasi = defuzzifikasi(hasil_agregasi)
    # menentukan status kelulusan
    prediksi_kelulusan = kelulusan(hasil_defuzzifikasi)
    # menghitung prediksi benar
    if int(prediksi_kelulusan) == int(aktual_kelulusan):
        prediksi_benar += 1
        print(f"Aktual: {aktual_kelulusan}, Prediksi: {prediksi_kelulusan}      :   B")
    else:
        print(f"Aktual: {aktual_kelulusan}, Prediksi: {prediksi_kelulusan}      :   SALAH")

# hitung akurasi
akurasi = (prediksi_benar / len(dataset)) * 100
print(f"\nAKURASI: {akurasi:.4f}%")
```
